# We will perform EDA

```python
# We will check the purchase distribution

plt.figure(figsize=(8,4))
sns.histplot(data=df, x='Purchase', bins=20, kde=True )
plt.xlabel('Purchase')
plt.ylabel('Frequency')
plt.title('Purchase Amount Distribution')
plt.show()
```

```python
# Check for outliers
# we will be detecting and checking for all the outliers in our dataset
# we will calculate quartiles and IQR

Q1 = df['Purchase'].quantile(0.25)
Q3 = df['Purchase'].quantile(0.75)
IQR = Q3-Q1

# we will define upper and lower bounds for the potential outliers

lower_bound= Q1 -1.5*IQR
upper_bound = Q3 +1.5*IQR

# now we will found the outliers

outliers = df[(df['Purchase']<lower_bound) | (df['Purchase']>upper_bound)]
outliers.head()
```

```python
#After detecting the outliers we will show the outliers in boxplot

plt.figure(figsize=(6,4))
sns.boxplot(data=df, y='Purchase')
plt.title('Outliers present in dataset')
plt.show()
```

```python
# Drop unnecessary fields
# We will be dropping product category_3 column as it has most nan values

df.drop(columns= 'Product_Category_3', inplace=True)
```

```python
df.columns
```

```python
#Convert categorical data into integer using map function (e.g 'Gender' column)
# In this dataframe we are going to change the gender column into integer 

gender_int = {'M' : 0 , 'F': 1}
df['Gender'] = df['Gender'].map(gender_int)
df.head()

# now you all can see that before in the gender field we had F & M and we can see that it's 1 & 0
```

```python
# Now we will be checking for the missing values in the dataframe 
#and after droping we will replace it with the mean of the column

missing_values = df.isnull().sum()
missing_values
```

```python
# as we can see that there are two columns in the dataframe which are having missing values and now we will replace the 
#missing values with their mean of the columns

mean_values = df['Product_Category_2'].mean()
df['Product_Category_2'].fillna(mean_values, inplace = True)
```

```python
mean_values1 = df['Product_Category_3'].mean()
df['Product_Category_3'].fillna(mean_values1, inplace=True)
```

```python
# Rename columns

df.columns
```

```python
# In this code we will change the name of the column: 'Gender' to 'Sex'

column_rename = {'Gender': 'Sex'}
df.rename(columns=column_rename, inplace=True)
df.head()
#as you can see that we have changed the column name from 'Gender' to 'Sex'
```

```python
# Map range variables into integers (e.g 'Age' column)
#as you can check in the table that in the age column we have ranges and we will be changing it into integers

age_mapping= {'0-17': 1,
             '18-25': 2,
             '26-35': 3,
             '36-50':4,
             '55+':5}

df['Age']=df['Age'].map(age_mapping)
df.tail()
```

# DATA VISUALIZATION

```python
# we will visualization every individual columns
```

```python
# Visualize numeric columns using histograms and box plots
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
for column in numeric_columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df, x=column, kde=True)
    plt.title(f'Histogram of {column}')
    plt.show()
    
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df, y=column)
    plt.title(f'Box Plot of {column}')
    plt.show()

# Visualize categorical columns using bar plots
categorical_columns = df.select_dtypes(include=['object']).columns
for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=column)
    plt.title(f'Count Plot of {column}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# We have to visualize the two columns i.e. Age vs Purchased

plt.figure(figsize=(8,4))
sns.boxplot(x='Age', y='Purchase',data=df)
plt.xlabel('Age')
plt.ylabel('Purchase Amount')
plt.title('Age vs Purchase')
plt.show()
```

```python
#Occupation vs Purchased
plt.figure(figsize=(8,4))
sns.boxplot(x='Occupation', y='Purchase',data=df)
plt.xlabel('Occupation')
plt.ylabel('Purchase Amount')
plt.title('Occupation vs Purchase')
plt.show()
```

```python
# Productcategory1 vs Purchased

plt.figure(figsize=(8,4))
sns.boxplot(x='Product_Category_1', y='Purchase',data=df)
plt.xlabel('Product_Category_1')
plt.ylabel('Purchase Amount')
plt.title('Product_Category_1 vs Purchase')
plt.show()
```

```python
# In this City category pie chart we will create a pie chart

city_counts = df['City_Category'].value_counts()
plt.figure(figsize=(8,4))
plt.pie(city_counts, labels=city_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('City Category Distribution')
plt.show()
```
